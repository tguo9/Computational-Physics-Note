## Demonstration of How openCV works 

## Parameters/Arguments:	
- ### image – 2Dnumpy array containing an image where objects are detected.
- ### scaleFactor – Parameter specifying how much the image size is reduced at each image scale.
- ### minNeighbors – Parameter specifying how many neighbors each candidate rectangle should have to retain it. [If too high -- high precion, but possibly also high false negatives (number of faces NOT detected); if too low, false positives.]
- ### minSize – Minimum possible object size. Objects smaller than that are ignored.
- ### maxSize – Maximum possible object size. Objects larger than that are ignored.
- ### flags – Parameter with the same meaning for an old cascade as in the function cvHaarDetectObjects. It is not used for a new cascade. [It's not always necessary to use this.]


## It returns:

- ### objects – Vector of rectangles where each rectangle contains the detected object: (x, y, w, h).  x and y are the lower left corner.


```python
'''
Face detection using openCV classifier   

'''

import cv2

# Get user supplied values
imagePath = 'FleetwoodMac.jpg'
#imagePath = 'Users/CareePath/xhuang/CompPhys-I/homework/FinalProject/whoswho.JPG'
cascPath = 'haarcascade_frontalface_default.xml'

# instantiate the haar cascade object
# This is basically a classifier:
# http://docs.opencv.org/2.4/modules/objdetect/doc/cascade_classification.html
faceCascade = cv2.CascadeClassifier(cascPath)

# Read the image
image = cv2.imread(imagePath)
print(image.shape)
# Detect faces in the image
# if small features are identified as faces, change scaleFactor to 1.2
# 
# if a face is "found" twice with two concentric boxes, set scaleFactor larger.
# Also watch what happens when minSize is set to (70, 70)
faces = faceCascade.detectMultiScale(
                                     image,
                                     scaleFactor=1.2,
                                     minNeighbors=5,
                                     minSize=(30, 30),
                                     flags = cv2.CASCADE_SCALE_IMAGE  # it seems to run w/o this line, too.
                                     )

print('faces type', type(faces))
# each entry in faces gives [x, y, w, h] 
print('Coordinates of faces:\n', faces)
print("Found {:d} faces!".format(len(faces)))

```

## Breakout Exercise:

### Draw Rectangles based on the coordinates contained in faces for the faces detected

```python
imagePath = 'FleetwoodMac.jpg'
image = cv2.imread(imagePath)

faces = faceCascade.detectMultiScale(
                                     image,
                                     scaleFactor=1.2,
                                     minNeighbors=5,
                                     minSize=(30, 30),
#                                      flags = cv2.CASCADE_SCALE_IMAGE  # it seems to run w/o this line, too.
                                     )

plt.figure(figsize = (8, 8))
plt.imshow(image)

for (x, y, w, h) in faces:
    
    plt.plot([x, x+w], [y, y], 'r-')
    plt.plot([x, x], [y, y+h], 'r-')
    plt.plot([x+w, x+w], [y, y+h], 'r-')
    plt.plot([x+w, x], [y+h, y+h], 'r-')

plt.grid('off')
plt.axis('off')
```
